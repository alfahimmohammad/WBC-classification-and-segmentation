```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
idg=ImageDataGenerator()
```

```python
training_data=idg.flow_from_directory('blood_cell_images/TRAIN',target_size=(64,64),class_mode='categorical',batch_size=32)
validation_data=idg.flow_from_directory('blood_cell_images/TEST',target_size=(64,64),class_mode='categorical',batch_size=32)
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
inc = InceptionV3(input_shape=(224,224,3), weights='imagenet', include_top=False)
```

```python
for i in inc.layers:
  i.trainable=False
```

```python
x=Flatten()(inc.output)
```

```python
pred=Dense(4,activation='softmax')(x)
```

```python
from keras.models import Model
 
```

```python
model=Model(inputs=inc.input,outputs=pred)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history_new=model.fit(training_data,validation_data=validation_data,epochs=30)
```
